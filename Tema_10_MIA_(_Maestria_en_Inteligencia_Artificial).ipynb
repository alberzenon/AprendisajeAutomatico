De acuerdo con Lubanovic, Python es un lenguaje dinámico, esto quiere decir que no es necesario declarar tipos de variables antes de usarse. Por ejemplo, si escribes x = 5, este tipo de lenguaje interpreta que 5 es un número entero y que, por lo tanto, la variable de x también lo es.

En los lenguajes dinámicos se puede lograr más con menos líneas de código, ya que no necesita ser compilado como otro tipo de lenguajes, sino que son interpretados por un programa llamado intérprete, sin embargo, también suelen ser más lentos. No obstante, su velocidad ha ido mejorando con el tiempo conforme sus intérpretes se han ido optimizando.

Por otra parte, todo en Python (valores booleanos, enteros, flotantes, cadenas, incluso grandes estructuras de datos, funciones y programas) se implementa como un objeto. Esto le da al lenguaje una consistencia (y características útiles) de las que carecen algunos otros
