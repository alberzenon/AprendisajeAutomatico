```python
import pandas as pd
```

```python
pacientes = pd.read_csv('https://raw.githubusercontent.com/alberzenon/AprendizajeAutomatico/main/Actividades/Arboles_de_decision/Pacientes2.csv' , engine = 'python', index_col=0)
```

```python
pacientes.head()
```

```python
pacientes.info()
```

```python
#variables predictoras
X=pacientes.iloc[:,1:11]

#variables a predecir
y= pacientes.iloc[:,0]


X.head()
```

```python
from sklearn.model_selection import train_test_split

#X_train y Y_train para entrenamiento
#X test y Y_test para prueba
#75% de los datos para el entrenamiento y el resto obtener la presisin del modelo
X_train, X_test, Y_train, Y_test = train_test_split(X, y, train_size=0.75, random_state=0)

```

```python
X_train.info()
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
#Llamamos al constructor del arbol de decision
arbol = DecisionTreeClassifier(max_depth=4)  

#
arbol_enfermedad = arbol.fit(X_train, Y_train)
```

```python
from matplotlib import pyplot as plt
from sklearn import tree

fig =plt.figure(figsize=(25,20))
tree.plot_tree(arbol_enfermedad, feature_names=(X.columns.values), class_names=list(y.values), filled=True)

plt.show()
```

```python
fig.savefig("Arbol de decision.png")
```

```python
#Prediccion con los datos  contenidos en la tabla de testing
Y_pred = arbol_enfermedad.predict(X_test)
Y_pred
```

```python
#Calculando la prcision del modelo
from sklearn.metrics import  confusion_matrix

Matriz_de_confucion = confusion_matrix(Y_test, Y_pred)

Matriz_de_confucion
```

```python
#Calculando la precision global del modelo

import numpy as np
```

```python
Precision_Global = np.sum(Matriz_de_confucion.diagonal()) / np.sum(Matriz_de_confucion)
Precision_Global
```

```python
#Precision por cada clase

Precision_No = ((Matriz_de_confucion[0,0])) / sum(Matriz_de_confucion[0,])
Precision_No
```

```python
Precision_Si = ((Matriz_de_confucion[1,1])) / sum(Matriz_de_confucion[1,])
Precision_Si
```
